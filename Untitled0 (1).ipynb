```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

try:
    # 1. VERİ YÜKLEME (Yeni master_data.csv dosyanı Colab'e yüklemeyi unutma)
    df = pd.read_csv('sample_data/master_data_1993_2024.csv')

    # Sütun isimlerini standartlaştıralım
    df.columns = df.columns.str.strip().str.lower()

    # Sayısal dönüşümler
    df['usdtry_q_avg'] = pd.to_numeric(df['usdtry_q_avg'], errors='coerce')
    df['uefa_country_coeff'] = pd.to_numeric(df['uefa_country_coeff'], errors='coerce')
    df['fifa_ranking'] = pd.to_numeric(df['fifa_ranking'], errors='coerce')

    # Milli Takım Başarı Endeksi (Sıralama düştükçe başarı artar, grafikte ters çeviriyoruz)
    df['fifa_success_index'] = df['fifa_ranking'].max() - df['fifa_ranking']

    # Zaman etiketi oluşturma (1993 Q2, 1993 Q4...)
    df['date_label'] = df['year'].astype(str) + " Q" + df['quarter'].astype(str)

    # 2. GÖRSELLEŞTİRME
    fig, ax1 = plt.subplots(figsize=(20, 10))

    # Dolar Kuru (Sol Eksen)
    sns.lineplot(x='date_label', y='usdtry_q_avg', data=df, ax=ax1, color='tab:blue', label='USD/TRY Rate', linewidth=3)
    ax1.set_ylabel('Exchange Rate (USD/TRY)', color='tab:blue', fontweight='bold')

    # Başarı Endeksleri (Sağ Eksen)
    ax2 = ax1.twinx()
    sns.lineplot(x='date_label', y='uefa_country_coeff', data=df, ax=ax2, color='tab:red', label='Club Success (UEFA)', linestyle='--', linewidth=2)
    sns.lineplot(x='date_label', y='fifa_success_index', data=df, ax=ax2, color='tab:green', label='National Team Success (FIFA Index)', linewidth=3)
    ax2.set_ylabel('Success Index', fontweight='bold')

    # Önemli Başarı Noktaları (Annotation)
    ax2.annotate('UEFA CUP 2000', xy=('2000 Q2', 26), xytext=('1997 Q1', 30),
                 arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, fontweight='bold')
    ax2.annotate('WORLD CUP 2002 (3rd)', xy=('2002 Q2', 40), xytext=('2003 Q4', 45),
                 arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, fontweight='bold')

    plt.title(f'Turkish Economy vs. Football Performance (1993-2024)\nTotal Data Points: {len(df)}', fontsize=16)

    # X eksenini okunabilir yapalım (Her 4 etikette bir göster)
    for i, label in enumerate(ax1.get_xticklabels()):
        if i % 4 != 0: label.set_visible(False)
    plt.xticks(rotation=45)

    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
    plt.show()

    # 3. HİPOTEZ TESTİ (Neşe Hoca'nın istediği asıl kısım)
    print("\n" + "="*50)
    print("🔬 PEARSON CORRELATION ANALYSIS (N=64)")
    print("="*50)

    r_uefa, p_uefa = pearsonr(df['usdtry_q_avg'], df['uefa_country_coeff'])
    r_fifa, p_fifa = pearsonr(df['usdtry_q_avg'], df['fifa_success_index'])

    print(f"Economy vs Club Success (UEFA): r={r_uefa:.3f}, p-value={p_uefa:.4f}")
    print(f"Economy vs National Team (FIFA): r={r_fifa:.3f}, p-value={p_fifa:.4f}")
    print("="*50)

except Exception as e:
    print(f"Hata: {e}")
```
